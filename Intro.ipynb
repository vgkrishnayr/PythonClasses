**Python Classes**

Python is an OOP language. Everything in Python is an object i.e., it is an object that belongs to a class..
An integer variable is an object instance of an integer class. Integer class provides methods to perform operations on object instance belonging to integer class.


```python
my_number = 7
print(type(my_number))
print(dir(my_number))
another_number = 5
print(my_number + another_number)  # __add__ method is invoked
print(my_number.__add__(another_number))
#
```
