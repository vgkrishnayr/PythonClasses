### Attributes
The features of the objects are represented using class attributes and instance attributes. Class attributes are shared by all instances and are set in the scope of Class, whereas instance attribute is specific/unique for an object instance and are set in the instance methods.

###  Instance methods
To set the instance attributes by default (initialize), an \_\_init__ instance method is used, this is often called as constructor method. To modify these instance attributes, we can have other instance methods. The instance methods have "self" as the first argument referring to the instance being created and other arguments are referred using "self". Self is the placeholder for the object. Instance methods have access to both instance attributes as well as class attributes.

```python
# class defintion
class Car:
  #class attribute
  num_of_wheels = 4

  #instance methods
  #instance initialization
  def __init__(self, color, mileage):
    #instance attributes
    self.color = color
    self.mileage = mileage

  def change_car_color(self, new_color):
    self.color = new_color
    return self.color

# instantiation of object
bmw = Car(color="Yellow", mileage=10000)
honda = Car(color="white", mileage=20000)


print(bmw.color, bmw.mileage, bmw.num_of_wheels)
print(honda.color, honda.mileage, honda.num_of_wheels)
print(Car.num_of_wheels)

print(bmw.change_car_color(new_color="Red"))
```

### Class methods

Class methods are bound to classes and have access to only class attributes (can be used to modify class attributes) and does not have access to instance attributes. To define class method, @classmethod decorator is used and also "cls" is the first argument indicating it is bound to class and other arguments are referred using cls. Instance methods are bound to objects and have access to both instance attributes and class attributes.



```python
# class defintion
class Car:
  #class attribute
  num_of_wheels = 4

  #instance methods
  #instance initialization
  def __init__(self, color, mileage):
    #instance attributes
    self.color = color
    self.mileage = mileage

  def change_car_color(self, new_color):
    self.color = new_color
    return self.color

  # Class method to modify class attributes  (Decorator- a function that calls another function, so classmethod() is an internal function)
  @classmethod
  def change_num_of_wheels(cls, new_num_of_wheels):
    cls.num_of_wheels = new_num_of_wheels
    return cls.num_of_wheels

# instantiation of object
bmw = Car(color="Yellow", mileage=10000)
print(Car.num_of_wheels)
print(Car.change_num_of_wheels(new_num_of_wheels=6))
print(Car.num_of_wheels)
```

### Static Methods
These are utility methods to do operations/actions with in the class and do not need access to class or instance attributes. (So we will not be passing self or cls as first argument)

```python
# class defintion
class Car:
  #class attribute
  num_of_wheels = 4

  #instance methods
  #instance initialization
  def __init__(self, color, mileage):
    #instance attributes
    self.color = color
    self.mileage = mileage

  def change_car_color(self, new_color):
    self.color = new_color
    return self.color

  # Class method to modify class attributes  (Decorator- a function that calls another function, so classmethod() is an internal function)
  @classmethod
  def change_num_of_wheels(cls, new_num_of_wheels):
    cls.num_of_wheels = new_num_of_wheels
    return cls.num_of_wheels

  @staticmethod
  def classify_car(car_oem):
    if car_oem == "bmw":
      return "German Car"
    elif car_oem == "honda":
      return "Japanese Car"
    else:
      return "Unknown"


# instantiation of object
bmw = Car(color="Yellow", mileage=10000)
print(Car.classify_car(car_oem="honda"))
print(bmw.classify_car(car_oem="bmw"))
```
