Python is an OOP language. Everything in Python is an object i.e., it is an object that belongs to a class..
An integer variable is an object instance of an integer class. Integer class provides methods to perform operations on object instance belonging to integer class.

**Object**: An object contains data and functionality together representing real-world objects/concepts. Functions correspond to operations representing real-world interactions of the objects.


```python
my_number = 7
print(type(my_number))
print(dir(my_number))
another_number = 5
print(my_number + another_number)  # __add__ method is invoked
print(my_number.__add__(another_number))
#
```

Similarly, we have string class and list class.

```python
#an instance of a string class
my_str = 'hello'
print(type(my_str))
print(dir(my_str)) #
print(my_str.upper(),"\n")

#an instance of a list class
my_list = [1, 2, 3]
print(type(my_list))
print(dir(my_list))
print("length of my list:", my_list.__len__())
my_list.append(4)
print(my_list, "\n","length of new list:", len(my_list))

```

## User defined classes

Classes are the blueprint for creating the objects. To create an object, class blueprint is defined and the objects are instantiated.

```python
# class defintion
class Car:
  pass

# instantiation of object
bmw = Car()
honda = Car()
#bmw and honda are two instances of Car class
print(bmw)
print(type(bmw))
print(dir(bmw))
```

The features of the objects are represented using class attributes and instance attributes. Class attributes are shared by all instances and are set in the scope of Class, whereas instance attribute is specific/unique for an object instance and are set in the instance methods. To set the instance attributes by default, an \_\_init__ instance method is used. To modify these instance attributes, we can have other instance methods.
